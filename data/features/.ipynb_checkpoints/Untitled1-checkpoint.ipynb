```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./pm_geo+rdf+kh.txt", sep="\s+", index_col=0)
```

```python
df
```

```python
random_idx = np.random.randint(0,100)
```

```python
l = df.iloc[random_idx]

EN = l[5:118]
H = l[118:118+113]
VDW = l[118+113:-1]
```

```python
from matplotlib import pyplot as plt
```

```python
x = np.linspace(2,30,113)
plt.plot(x, EN, label="EN")
plt.plot(x, H, label="hardness")
plt.plot(x, VDW, label="VDW")
plt.legend()
plt.show()
```
